```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Visualize the data with a line plot
plt.plot(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dummy Data - Line Plot')
plt.show()

# Generate more complex data for heatmap
np.random.seed(0)
data = np.random.rand(10, 10)

# Create a DataFrame for heatmap
df_heatmap = pd.DataFrame(data)

# Visualize the data with a heatmap
plt.imshow(df_heatmap, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Dummy Data - Heatmap')
plt.show()
```
